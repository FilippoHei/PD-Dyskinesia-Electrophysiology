```python
PATH_CUR    = os.path.abspath(os.curdir)                 # current code
PATH        = (str(Path(PATH_CUR).parent))               # data repository: upper directory where datasets situated
SUB_LIST    = utils_misc.get_SUB_list(DATA_IO.path_data) # get the SUB id list which we have a recording of them
fs          = 2048
```

```python
dataset                                         = pd.read_pickle(DATA_IO.path_events + "LFP_TAPPING_EVENTS_MOTOR_AREA.pkl")

df_TAPPING                                      = {} 
df_TAPPING["Motor"]                             = {}
df_TAPPING["Motor"]["controlateral"]            = {}
df_TAPPING["Motor"]["controlateral"]["LID_no"]  = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_arm == "none")]
df_TAPPING["Motor"]["controlateral"]["LID_all"] = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_arm != "none")]

df_TAPPING["Motor"]["controlateral"]["LID_arm_mild"]       = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_arm == "mild")]
df_TAPPING["Motor"]["controlateral"]["LID_arm_moderate"]   = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_arm == "moderate")]
df_TAPPING["Motor"]["controlateral"]["LID_arm_severe"]     = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_arm == "severe")]

df_TAPPING["Motor"]["controlateral"]["LID_total_mild"]     = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_total == "mild")]
df_TAPPING["Motor"]["controlateral"]["LID_total_moderate"] = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_total == "moderate")]
df_TAPPING["Motor"]["controlateral"]["LID_total_severe"]   = dataset[(dataset.event_laterality != dataset.LFP_hemisphere) & (dataset.dyskinesia_total == "severe")]

df_TAPPING["Motor"]["controlateral"]["LID_no"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_all"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_arm_mild"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_arm_moderate"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_arm_severe"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_total_mild"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_total_moderate"].reset_index(drop=True, inplace=True)
df_TAPPING["Motor"]["controlateral"]["LID_total_severe"].reset_index(drop=True, inplace=True)
```
