# 2. Spectogram

```python
def get_mean_LFP_activity(dataset, fs):
    
    mean_LFP = []
    
    for i in range(len(dataset)):
        try:
            mean_LFP.append(np.nan_to_num(LFP.get_onset_aligned_ephys_recording(dataset, fs=fs, index=i)))
        except:
            pass
            
    return np.mean(mean_LFP, axis=0)
```
