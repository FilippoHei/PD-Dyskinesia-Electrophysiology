```python
# Import public packages and functions
import os
import pandas as pd
import numpy as np
import sys
import json
from pathlib import Path
import seaborn as sns

# inserting the lib folder to the compiler
sys.path.insert(0, './lib')
sys.path.insert(0, './utils/')

import utils_plotting
```

```python
SUB = 16
```

```python
def print_event_distributions(dataset, SUB):

    events_left      = dataset[(dataset.patient == SUB) & (dataset.hemisphere == "left")]
    events_right     = dataset[(dataset.patient == SUB) & (dataset.hemisphere == "right")]
    events_bilateral = dataset[(dataset.patient == SUB) & (dataset.hemisphere == "bilateral")]

    print(" ---------------------------------------------------------------------------")
    print(" SUB - " + str(SUB))
    print(" ---------------------------------------------------------------------------")
    print(" Event Distribution ")
    
    print("     > left hemisphere [N=" + str(len(events_left))+"]")
    print("         > voluntary events [N="+str(len(events_left[events_left.is_voluntary==True]))+"]")
    print("             - voluntary tapping : "+str(round((len(events_left[events_left.event_category=="voluntary_tapping"])/len(events_left)*100),2)) +"%")
    print("         > involuntary events [N="+str(len(events_left[events_left.is_voluntary==False]))+"]")
    print("             - involuntary tapping  : "+str(round((len(events_left[events_left.event_category=="involuntary_tapping"])/len(events_left)*100),2)) +"%")
    print("             - involuntary movement : "+str(round((len(events_left[events_left.event_category=="involuntary_movement"])/len(events_left)*100),2)) +"%")
    
    print("     > right hemisphere [N=" + str(len(events_right))+"]")
    print("         > voluntary events [N="+str(len(events_right[events_right.is_voluntary==True]))+"]")
    print("             - voluntary tapping : "+str(round((len(events_right[events_right.is_voluntary==True])/len(events_right)*100),2)) +"%")
    print("         > involuntary events [N="+str(len(events_right[events_right.is_voluntary==False]))+"]")
    print("             - involuntary tapping  : "+str(round((len(events_right[events_right.event_category=="involuntary_tapping"])/len(events_right)*100),2)) +"%")
    print("             - involuntary movement : "+str(round((len(events_right[events_right.event_category=="involuntary_movement"])/len(events_right)*100),2)) +"%")
    
    print("     > bilateral [N=" + str(len(events_bilateral))+"]")
    print("         > voluntary events [N="+str(len(events_bilateral[events_bilateral.is_voluntary==True]))+"]")
    print("             - voluntary tapping : "+str(round((len(events_bilateral[events_bilateral.is_voluntary==True])/len(events_bilateral)*100),2)) +"%")
    print("         > involuntary events [N="+str(len(events_bilateral[events_bilateral.is_voluntary==False]))+"]")
    print("             - involuntary tapping  : "+str(round((len(events_bilateral[events_bilateral.event_category=="involuntary_tapping"])/len(events_bilateral)*100),2)) +"%")
    print("             - involuntary movement : "+str(round((len(events_bilateral[events_bilateral.event_category=="involuntary_movement"])/len(events_bilateral)*100),2)) +"%")
```

# 1. Load Event Dataframe of The Patient

```python
df_events_016 = pd.read_csv("events/SUB_016_EVENTS.csv")
```

# 2. Voluntary vs Non-Voluntary Events

```python
events = df_events_016.copy()
events
```

```python
plt        = utils_plotting.get_figure_template()

dist_type  = plt.subplot2grid((75, 50), (0, 0), colspan=8, rowspan=7)
dist_hem   = plt.subplot2grid((75, 50), (0, 14), colspan=8, rowspan=7)
dist_task  = plt.subplot2grid((75, 50), (0, 28), colspan=8, rowspan=7)
dist_cat   = plt.subplot2grid((75, 50), (0, 42), colspan=8, rowspan=7)

dist_move_dys = plt.subplot2grid((75, 50), (15, 0), colspan=8, rowspan=7)
dist_tap_dys  = plt.subplot2grid((75, 50), (15, 14), colspan=8, rowspan=7)
dist_vol_dys  = plt.subplot2grid((75, 50), (15, 28), colspan=8, rowspan=7)
dist_cat_dys  = plt.subplot2grid((75, 50), (15, 42), colspan=8, rowspan=7)

dist_dur_type = plt.subplot2grid((75, 50), (30, 0), colspan=8, rowspan=10)
dist_dur_vol  = plt.subplot2grid((75, 50), (30, 14), colspan=8, rowspan=10)
dist_dur_task = plt.subplot2grid((75, 50), (30, 28), colspan=8, rowspan=10)
dist_dur_hem  = plt.subplot2grid((75, 50), (30, 42), colspan=8, rowspan=10)

###############################################################################################################

dist_type  = sns.countplot(events, x="hemisphere", hue="event", orient="h", ax=dist_type)
dist_type.set_xlabel("")
dist_type.set_title("event type", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_type.legend(["move","tap"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                 bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_hem = sns.countplot(events, x="hemisphere", hue="is_voluntary", orient="h", ax=dist_hem)
dist_hem.set_xlabel("")
dist_hem.set_title("event volition", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_hem.legend(["involuntary","voluntary"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                 bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_task = sns.countplot(events, x="event", hue="task", orient="h", ax=dist_task)
dist_task.set_xlabel("")
dist_task.set_title("event per task", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_task.legend(["rest task","tapping task"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                 bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_cat = sns.countplot(events, x="hemisphere", hue="event_category", orient="h", ax=dist_cat)
dist_cat.set_xlabel("")
dist_cat.set_title("event category per hemisphere", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_cat.legend(["move","inv. tap","vol. tap"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                 bbox_to_anchor=(0.5, -0.75), title=None, ncol=2, frameon=False)

utils_plotting.set_axis(dist_type)
utils_plotting.set_axis(dist_hem)
utils_plotting.set_axis(dist_task)
utils_plotting.set_axis(dist_cat)

###############################################################################################################

dist_move_dys = sns.countplot(events[events.event=="move"], x="hemisphere", hue="dyskinesia_score", orient="h", ax=dist_move_dys)
dist_move_dys.set_xlabel("")
dist_move_dys.set_title("move per dyskinesia severity", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_move_dys.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_tap_dys = sns.countplot(events[events.event=="tap"], x="hemisphere", hue="dyskinesia_score", orient="h", ax=dist_tap_dys)
dist_tap_dys.set_xlabel("")
dist_tap_dys.set_title("tapping per dyskinesia severity", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_tap_dys.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_vol_dys = sns.countplot(events, x="is_voluntary", hue="dyskinesia_score", orient="h", ax=dist_vol_dys)
dist_vol_dys.set_xlabel("")
dist_vol_dys.set_xticklabels(["involuntary","voluntary"])
dist_vol_dys.set_title("event volition per dyskinesia severity", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_vol_dys.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

dist_cat_dys = sns.countplot(events, x="event_category", hue="dyskinesia_score", orient="h", ax=dist_cat_dys)
dist_cat_dys.set_xlabel("")
dist_cat_dys.set_xticklabels(["move", "inv. tap", "vol. tap"])
dist_cat_dys.set_title("event category per dyskinesia severity", fontsize=utils_plotting.LABEL_SIZE, weight="bold")
dist_cat_dys.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)

utils_plotting.set_axis(dist_move_dys)
utils_plotting.set_axis(dist_tap_dys)
utils_plotting.set_axis(dist_vol_dys)
utils_plotting.set_axis(dist_cat_dys)

###############################################################################################################

dist_dur_type = sns.boxplot(data=events, x="event", y="duration", hue="dyskinesia_score", showfliers=False, ax=dist_dur_type)
dist_dur_type.set_xlabel("")
dist_dur_type.set_ylabel("event duration [sec]", fontsize=utils_plotting.LABEL_SIZE)
dist_dur_type.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)
dist_dur_type.set_title("duration for event type", fontsize=utils_plotting.LABEL_SIZE, weight="bold")

dist_dur_vol = sns.boxplot(data=events, x="is_voluntary", y="duration", hue="dyskinesia_score", showfliers=False, ax=dist_dur_vol)
dist_dur_vol.set_xlabel("")
dist_dur_vol.set_xticklabels(["involuntary","voluntary"])
dist_dur_vol.set_ylabel("event duration [sec]", fontsize=utils_plotting.LABEL_SIZE)
dist_dur_vol.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)
dist_dur_vol.set_title("duration for volition of event", fontsize=utils_plotting.LABEL_SIZE, weight="bold")

dist_dur_task = sns.boxplot(data=events, x="task", y="duration", hue="dyskinesia_score", showfliers=False, ax=dist_dur_task)
dist_dur_task.set_xlabel("")
dist_dur_task.set_ylabel("event duration [sec]", fontsize=utils_plotting.LABEL_SIZE)
dist_dur_task.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)
dist_dur_task.set_title("duration for task", fontsize=utils_plotting.LABEL_SIZE, weight="bold")

dist_dur_hem = sns.boxplot(data=events, x="hemisphere", y="duration", hue="dyskinesia_score", showfliers=False, ax=dist_dur_hem)
dist_dur_hem.set_xlabel("")
dist_dur_hem.set_ylabel("event duration [sec]", fontsize=utils_plotting.LABEL_SIZE)
dist_dur_hem.legend(["dys=0","dys=1"], loc="lower center", fontsize=utils_plotting.LABEL_SIZE, 
                       bbox_to_anchor=(0.5, -0.6), title=None, ncol=2, frameon=False)
dist_dur_hem.set_title("duration for hemispheres", fontsize=utils_plotting.LABEL_SIZE, weight="bold")

utils_plotting.set_axis(dist_dur_type)
utils_plotting.set_axis(dist_dur_vol)
utils_plotting.set_axis(dist_dur_task)
utils_plotting.set_axis(dist_dur_hem)

plt.savefig("figures/SUB" + str(SUB) +"_event_stats.svg", dpi=300)
```
