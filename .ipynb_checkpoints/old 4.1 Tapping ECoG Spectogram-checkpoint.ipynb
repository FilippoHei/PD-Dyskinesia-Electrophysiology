```python
def get_mean_ECoG_activity(dataset, fs):
    
    mean_LFP = []
    
    for i in range(len(dataset)):
        try:
            mean_LFP.append(np.nan_to_num(LFP.get_onset_aligned_ephys_recording(dataset, fs=fs, index=i)))
        except:
            pass
            
    return np.mean(mean_LFP, axis=0)
```

### 2.1.1 LID vs No-LID State

```python
from stockwell import st
def spectogram_stockwell(t, recording, fmin, fmax, color, figure_name):

    """
    Description
        It is a technique based on continious wavelet transformation by utilizing gaussian window to get 
        time-frequency representation of signals.

        Github : https://github.com/claudiodsf/stockwell
        Paper  : 10.1109/78.492555
        
    Parameters:
        t               : 1xN float array, containing the time instances of recording
        recording       : 1xN float array, containing the values of recording
        fmin            : float, the lower bound of frequency band of interest [Hz]
        fmax            : float, the upper bound of frequency band of interest [Hz]
    """
    
    df           = 1/(t[-1]-t[0])  # sampling step in frequency domain (Hz)
    fmin_samples = int(fmin/df)
    fmax_samples = int(fmax/df)
    stock        = st.st(recording, fmin_samples, fmax_samples)
    extent       = (t[0], t[-1], fmin, fmax)
    
    plt   = utils_plotting.get_figure_template()
    ax1   = plt.subplot2grid((75, 40), (0, 0) , colspan=40, rowspan=10)
    ax2   = plt.subplot2grid((75, 40), (11, 0) , colspan=40, rowspan=25)

    # plotting
    ax1.plot(t, recording, linewidth=0.5, color=color)
    ax1.set(ylabel='amplitude')
    ax1.set_yticklabels([])
    ax1.set_xticklabels([])
    ax1.set_xlim([-2,2])
    ax1.axvline(x=0, ymin=-1, ymax=1, ls='-', color="dimgrey")

    ax2.imshow(np.abs(stock), origin='lower', extent=extent, vmin=1.e-8, vmax=1.e-6)
    ax2.axis('tight')
    ax2.set(xlabel='time (s)', ylabel='frequency (Hz)')
    ax2.set_ylim([fmin,fmax])

    utils_plotting.set_axis(ax1)
    utils_plotting.set_axis(ax2)

    plt.savefig(figure_name + ".png", dpi=300)
    plt.savefig(figure_name + ".svg", dpi=300)
```

```python
spectogram_stockwell(time_vector, noLID, fmin=4, fmax=100, 
                                               color=utils_plotting.colors["no_LID"],
                                               figure_name=DATA_IO.path_figure + "ECoG/ECoG_spectogram_noLID")
```

```python
mwt = compute_wavelet_transform(noLID, fs=fs, n_cycles=7, freqs=np.linspace(4, 100, 50))
plot_timefrequency(time_vector, np.linspace(4, 100, 50), mwt)
```
