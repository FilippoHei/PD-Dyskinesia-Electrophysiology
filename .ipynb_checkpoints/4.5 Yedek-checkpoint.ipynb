## methods

```python
from scipy.spatial import cKDTree
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import mutual_info_score

def mutual_information_with_permutation_test(dataset, feature_column, target_column, num_permutations=1000):
    # put seed for reproducibility
    np.random.seed(100)  # Set the seed
    
    # Calculate observed mutual information
    observed_mi = mutual_info_score(dataset[feature_column], dataset[target_column])

    # Store the mutual information values from permutations
    permuted_mi = []

    for _ in range(num_permutations):
        # Shuffle the target variable
        shuffled_target   = np.random.permutation(dataset[target_column])
        
        # Calculate mutual information for the shuffled data
        permuted_mi_value = mutual_info_score(dataset[feature_column], shuffled_target)
        permuted_mi.append(permuted_mi_value)

    # Calculate the p-value
    p_value = np.sum(np.array(permuted_mi) >= observed_mi) / num_permutations
    return observed_mi, p_value
    


def create_spatial_neighboorhoods_on_cortical_surface(df_activity, radius):
        # Step 1: Calculate distances
    coordinates = df_activity[['x_mapped', 'y_mapped', 'z_mapped']].values
    
    # Use Nearest Neighbors to find neighbors within a certain radius
    neighbors = NearestNeighbors(radius=radius, algorithm="kd_tree")  # Adjust radius as needed
    neighbors.fit(coordinates)
    
    # Find neighbors for each point
    distances, indices = neighbors.radius_neighbors(coordinates)
    
    # Step 2: Create neighborhoods and assign numeric codes
    neighborhood_codes = {}
    code_counter = 0
    
    for i, (dist, idx) in enumerate(zip(distances, indices)):
        if i not in neighborhood_codes:  # Check if already assigned
            # Assign a new neighborhood code
            neighborhood_codes[i] = code_counter
            code_counter += 1
            
        # Assign the same code to all neighbors
        for neighbor in idx:
            if neighbor not in neighborhood_codes:
                neighborhood_codes[neighbor] = neighborhood_codes[i]  # Same code as current point
    
    # Map the codes back to the original DataFrame
    df_activity['neighbourhood_id'] = df_activity.index.map(neighborhood_codes)
    
    # Step 3: Calculate centroids for each neighborhood
    centroid_dict = {}
    for code in set(neighborhood_codes.values()):
        # Get all coordinates for the current neighborhood code
        coords_in_neighborhood = df_activity[df_activity['neighbourhood_id'] == code][['x', 'y', 'z']]
        
        # Calculate the centroid (mean) coordinates
        centroid = coords_in_neighborhood.mean().to_dict()  # Convert to dictionary
        centroid_dict[code] = centroid
    return df_activity, centroid_dict
```

```python
cortex_mesh   = utils_io.load_cortical_atlas_meshes()
cortex_right  = cortex_mesh["right_hemisphere"]
cortex_left   = cortex_mesh["left_hemisphere"]
df            = pd.DataFrame(ECOG_feature)
df            = map_ECOG_activity_dataframe_to_surface(df, cortex_right)
df, centroids = create_spatial_neighboorhoods_on_cortical_surface(df, radius=8)
```
