```python
# Import public packages and functions
import os
import pandas as pd
import numpy as np
import sys
import json
from pathlib import Path
import pyvista as pv

import seaborn as sns
import matplotlib.pyplot as plt
import scikit_posthocs as sp

import warnings
warnings.filterwarnings("ignore")

# inserting the lib folder to the compiler
sys.path.insert(0, './lib')
sys.path.insert(0, './utils/')

import utils_io, utils_misc, utils_spatial_analysis, utils_correlation, utils_statistics, utils_plotting, utils_plotting_correlation, utils_plotting_spatial_mapping

from lib_data import DATA_IO
from lib_Subcortical_Atlases import Distal_Atlas
```

# 1. Create STN Grid and Assign LFP Channels to Grid Cells

```python
# load LFP-PSD dataframe
LFP_PSD_controlateral = utils_io.load_LFP_event_PSD(event_category="tapping", event_laterality="controlateral") 

# concat all different dyskinesia severity dataframe into single dataframe
LFP_PSD = pd.DataFrame()
for severity in LFP_PSD_controlateral.keys():
    LFP_PSD = pd.concat([LFP_PSD, LFP_PSD_controlateral[severity]], ignore_index=True)

LFP_PSD["severity"] = LFP_PSD['dyskinesia_arm']

# Based on event onset time, assign no-LID events into noLID_noDOPA and noLID_DOPA groups
LFP_PSD.loc[(LFP_PSD['event_start_time'] < 30)  & (LFP_PSD['severity'] == "none"), 'severity'] = 'noLID_noDOPA'
LFP_PSD.loc[(LFP_PSD['event_start_time'] >= 30) & (LFP_PSD['severity'] == "none"), 'severity'] = 'noLID_DOPA'

# arrange the dataframe
LFP_PSD['hemisphere'] = LFP_PSD['LFP_hemisphere']
LFP_PSD['channel']    = LFP_PSD['LFP_channel']
LFP_PSD = LFP_PSD[['patient', 'hemisphere', 'channel', 
                   'pre_event_theta_mean', 'pre_event_alpha_mean', 
                   'pre_event_beta_low_mean','pre_event_beta_high_mean','pre_event_gamma_mean',
                   'event_theta_mean', 'event_alpha_mean', 
                   'event_beta_low_mean','event_beta_high_mean','event_gamma_mean', 
                   'post_event_theta_mean', 'post_event_alpha_mean',
                   'post_event_beta_low_mean','post_event_beta_high_mean','post_event_gamma_mean',
                   'severity']]

features = ['pre_event_theta_mean', 'pre_event_alpha_mean', 'pre_event_beta_low_mean',
            'pre_event_beta_high_mean','pre_event_gamma_mean',
            'event_theta_mean', 'event_alpha_mean', 'event_beta_low_mean',
            'event_beta_high_mean','event_gamma_mean', 
            'post_event_theta_mean', 'post_event_alpha_mean',
            'post_event_beta_low_mean','post_event_beta_high_mean','post_event_gamma_mean']

# load STN mesh
STN_mesh         = utils_io.load_STN_meshes() 
# load LFP channels MNI coordinates
MNI_LFP_channels = pd.read_pickle(DATA_IO.path_coordinates + "MNI_LFP_motor_channels.pkl") 
# create a 3D grid around STN
n_bins           = 4
STN_grid         = utils_spatial_analysis.create_3D_grid_around_anatomical_structure(STN_mesh["right_hemisphere"], n_bins=n_bins)
# find the position of each channel in these 3D grid across x,y,z axes
MNI_LFP_channels = utils_spatial_analysis.assign_recording_channels_to_grid_cells(MNI_LFP_channels, STN_grid)

# Merge information of STN grid cells into the LFP_PSD dataframe
# merge the data frames on 'patient', 'hemisphere', and 'channel'
LFP_PSD = pd.merge(LFP_PSD, 
                   MNI_LFP_channels[['patient', 'hemisphere', 'channel', 'grid_bin_x', 'grid_bin_y', 'grid_bin_z']],
                   on=['patient', 'hemisphere', 'channel'], 
                   how='left')
LFP_PSD = LFP_PSD.dropna(subset=['grid_bin_x', 'grid_bin_y', 'grid_bin_z'])
LFP_PSD["severity_numeric"] = LFP_PSD['severity'].map({'noLID_noDOPA':0, 'noLID_DOPA':1, 'mild':2, 'moderate':3})
LFP_PSD.reset_index(drop=True, inplace=True)
```

# 2. Measure the Gridwise Linear Relationship (Neural Features & Dyskinesia Severiy)

## 2.1 Measure Correlation for Each Bin

```python
corr_x = utils_correlation.spearsman_correlation_along_spatial_axes(pd.DataFrame(LFP_PSD), features, target_feature="severity_numeric", 
                                                                    axis="x", n_bins=n_bins, patient_grouping=False)
corr_y = utils_correlation.spearsman_correlation_along_spatial_axes(pd.DataFrame(LFP_PSD), features, target_feature="severity_numeric", 
                                                                    axis="y", n_bins=n_bins, patient_grouping=False)
corr_z = utils_correlation.spearsman_correlation_along_spatial_axes(pd.DataFrame(LFP_PSD), features, target_feature="severity_numeric", 
                                                                    axis="z", n_bins=n_bins, patient_grouping=False)

# apply grid-bin based multiple comparison correction which means that, for each selected bin for an particular axis,
# we corrected p-values based on correlation calculation made only in this grid bin.
corr_x['pvalue_corrected'] = utils_statistics.apply_multiple_correction(corr_x['pvalue'].values, 'holm')
corr_y['pvalue_corrected'] = utils_statistics.apply_multiple_correction(corr_y['pvalue'].values, 'holm')
corr_z['pvalue_corrected'] = utils_statistics.apply_multiple_correction(corr_z['pvalue'].values, 'holm')

corr_x['significant']      = corr_x['pvalue_corrected'] <= 0.05
corr_y['significant']      = corr_y['pvalue_corrected'] <= 0.05
corr_z['significant']      = corr_z['pvalue_corrected'] <= 0.05

features_order = ['pre_event_theta_mean', 'event_theta_mean', 'post_event_theta_mean',
                  'pre_event_beta_low_mean', 'event_beta_low_mean', 'post_event_beta_low_mean',
                  'pre_event_beta_high_mean', 'event_beta_high_mean', 'post_event_beta_high_mean',
                  'pre_event_gamma_mean', 'event_gamma_mean', 'post_event_gamma_mean']
```

## 2.2 Plot Correlation for Each Bin

```python
plt     = utils_plotting.get_figure_template()
axis_x  = plt.subplot2grid((77, 66), (0, 0) , colspan=32, rowspan=10)
axis_x  = utils_plotting_correlation.plot_grid_wise_correlation(corr_x, features_order, n_bins=n_bins, axis=axis_x)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_corr_x.svg", dpi=300)

plt     = utils_plotting.get_figure_template()
axis_y  = plt.subplot2grid((77, 66), (0, 0) , colspan=32, rowspan=10)
axis_y  = utils_plotting_correlation.plot_grid_wise_correlation(corr_y, features_order, n_bins=n_bins, axis=axis_y)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_corr_y.svg", dpi=300)

plt     = utils_plotting.get_figure_template()
axis_z  = plt.subplot2grid((77, 66), (0, 0) , colspan=32, rowspan=10)
axis_z  = utils_plotting_correlation.plot_grid_wise_correlation(corr_z, features_order, n_bins=n_bins, axis=axis_z)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_corr_z.svg", dpi=300)
```

## 2.3 Measure Adjusted R2 for Each Bin    

```python
predictors = ['pre_event_theta_mean', 'event_theta_mean', 'post_event_theta_mean',
              'pre_event_beta_low_mean', 'event_beta_low_mean', 'post_event_beta_low_mean',
              'pre_event_beta_high_mean', 'event_beta_high_mean', 'post_event_beta_high_mean',
              'pre_event_gamma_mean', 'event_gamma_mean', 'post_event_gamma_mean']

df_x_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "x")
df_y_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "y")
df_z_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "z")
```

```python
predictors = ['pre_event_theta_mean', 'event_theta_mean', 'post_event_theta_mean']

df_x_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "x")
df_y_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "y")
df_z_r2    = utils_statistics.measure_adjusted_r2_in_grid_bins_along_axis(LFP_PSD, predictors, n_bins, "z")
```

## 2.4 Plot Adjusted R2 for Each Bin    

```python
def plot_adjusted_R2_in_grid_bins_along_axis(df_adj_r2, n_bins, axis):
    axis.scatter(df_adj_r2.grid_bin, df_adj_r2.adjusted_r2, c=df_adj_r2.adjusted_r2, cmap='summer', vmin=0, vmax=0.5, s=10)
    axis.set_xlim([-1,n_bins])
    axis.set_ylim([-0.1,0.6])
    
    axis.set_xlabel("")
    axis.set_ylabel("")
    
    axis.spines.left.set_visible(False)
    axis.spines.bottom.set_visible(False)
    axis.spines.right.set_visible(False)
    axis.spines.top.set_visible(False)
    axis.xaxis.set_ticks_position('none')  # Hide the x-axis ticks
    axis.xaxis.set_ticks_position('none')  # Hide the x-axis ticks
    
    axis.set_xticklabels(axis.get_xticklabels(), fontsize=utils_plotting.LABEL_SIZE)
    axis.set_yticklabels(axis.get_yticklabels(), fontsize=utils_plotting.LABEL_SIZE)
```

```python
plt  = utils_plotting.get_figure_template()
axis = plt.subplot2grid((77, 66), (0, 0) , colspan=10, rowspan=5)
axis = plot_adjusted_R2_in_grid_bins_along_axis(df_x_r2, n_bins, axis)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_adj_r2_x.svg", dpi=300)

plt  = utils_plotting.get_figure_template()
axis = plt.subplot2grid((77, 66), (0, 0) , colspan=10, rowspan=5)
axis = plot_adjusted_R2_in_grid_bins_along_axis(df_y_r2, n_bins, axis)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_adj_r2_y.svg", dpi=300)

plt  = utils_plotting.get_figure_template()
axis = plt.subplot2grid((77, 66), (0, 0) , colspan=10, rowspan=5)
axis = plot_adjusted_R2_in_grid_bins_along_axis(df_z_r2,  n_bins, axis)
plt.savefig(DATA_IO.path_figure + "correlation/LFP/grid_wise_adj_r2_z.svg", dpi=300)
```

# 3. Measure & Plot the Adjusted R2 for Each Cell

```python
df_R2_theta     = utils_statistics.measure_adjusted_r2_in_grid_cells_for_frequency_bands(pd.DataFrame(LFP_PSD), n_bins=n_bins,
                                                                                         frequency_band="theta", target_feature="severity_numeric")
df_R2_beta_low  = utils_statistics.measure_adjusted_r2_in_grid_cells_for_frequency_bands(pd.DataFrame(LFP_PSD), n_bins=n_bins,
                                                                                         frequency_band="beta_low", target_feature="severity_numeric")
df_R2_beta_high = utils_statistics.measure_adjusted_r2_in_grid_cells_for_frequency_bands(pd.DataFrame(LFP_PSD), n_bins=n_bins,
                                                                                         frequency_band="beta_high", target_feature="severity_numeric")
df_R2_gamma     = utils_statistics.measure_adjusted_r2_in_grid_cells_for_frequency_bands(pd.DataFrame(LFP_PSD), n_bins=n_bins,
                                                                                         frequency_band="gamma", target_feature="severity_numeric")
```

```python
grid_bin_center_coordianates = utils_spatial_analysis.extract_grid_cell_centers(STN_grid, n_bins)
STN_dynamics                 = pd.DataFrame(grid_bin_center_coordianates)

STN_dynamics                 = STN_dynamics.merge(df_R2_theta, left_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], right_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], how='left')
STN_dynamics['R2_theta']     = STN_dynamics['adjusted_r2']
STN_dynamics                 = STN_dynamics.drop(columns=['adjusted_r2','frequency_band'])

STN_dynamics                 = STN_dynamics.merge(df_R2_beta_low, left_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], right_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], how='left')
STN_dynamics['R2_beta_low']  = STN_dynamics['adjusted_r2']
STN_dynamics                 = STN_dynamics.drop(columns=['adjusted_r2','frequency_band'])

STN_dynamics                 = STN_dynamics.merge(df_R2_beta_high, left_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], right_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], how='left')
STN_dynamics['R2_beta_high'] = STN_dynamics['adjusted_r2']
STN_dynamics                 = STN_dynamics.drop(columns=['adjusted_r2','frequency_band'])

STN_dynamics                 = STN_dynamics.merge(df_R2_gamma, left_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], right_on=['grid_bin_x', 'grid_bin_y','grid_bin_z'], how='left')
STN_dynamics['R2_gamma']     = STN_dynamics['adjusted_r2']
STN_dynamics                 = STN_dynamics.drop(columns=['adjusted_r2','frequency_band'])

STN_dynamics["x"]            = STN_dynamics["grid_bin_center_x"]
STN_dynamics["y"]            = STN_dynamics["grid_bin_center_y"]
STN_dynamics["z"]            = STN_dynamics["grid_bin_center_z"]

STN_dynamics.replace(np.nan, 0, inplace=True)

STN_dynamics                 = STN_dynamics.loc[~(STN_dynamics[['R2_theta', 'R2_beta_low', 'R2_beta_high', 'R2_gamma']] == 0).all(axis=1)]
```

```python
stn_activity_mesh = utils_plotting_spatial_mapping.map_electrophysiological_activity_to_anatomical_surface(STN_dynamics, 
                                                                                                           STN_mesh["right_hemisphere"], 
                                                                                                           function='gaussian', 
                                                                                                           epsilon=2)
plotter           = utils_plotting_spatial_mapping.plot_LFP_activity_distribution(STN_mesh["right_hemisphere"], stn_activity_mesh, feature="theta", 
                                                                                  cmap="Reds", clim=[0.0, 0.3], file_path=DATA_IO.path_figure + "LFP_Maps/")
```
